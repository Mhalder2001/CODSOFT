```python
from  sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=data.drop('Class',axis=1)
y=data['Class'].ravel()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
len(x_train),len(x_test),len(y_test),len(y_train)
```

```python
Lreg=LogisticRegression(random_state=22)
Lreg.fit(x_train,y_train.ravel())
```

```python
y_predict=Lreg.predict(x_test).ravel()
y_predict
```

```python
data.Class
```

```python
Lreg.score(x_test,y_test)   #our model is good fitted
 
```

```python
## create confussion matrix
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_predict)
score
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,y_predict).ravel()
print("True Negative",tn)
print("False positive",fp)
print("False Negative",fn)
print("True positive",tp)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

#Recall gives us an idea about when it actually yes ,how often does it predict yes
#here recall value with respect to 1 is called sensitivity ,, recall value with respect to 0 is called specificity.
#the diagonal value of the classification report is true value i.e True Positive(TP) and True Negative(TN) 
#Precision tells us about when it predict yes how often correct
1.precision=TP/(FP+TP)
2.ACCURACY=TP+FN/(TP+FN+TN+FP)
