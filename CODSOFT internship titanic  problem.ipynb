#DATA CLEANING

```python
average_age=data['Age'].mean()
data["Age"].fillna(average_age,inplace=True)
```

```python
pd.isnull(data['Age']).sum()  ##we remove the missing value of age column
```

```python
import statsmodels.api as sm
```

```python
print(data.head())
```

```python
data['Sex']=data['Sex'].map({'male':1,'female':0})
data    ## convert non-numeric to numeric column
```

Here the dependent/response variable is catgrorical variable i.e means this take only binnary data ie 0,1 so we fit logistic regression model

```python
x=data[['Pclass','Sex','Age']]
y=data['Survived']
x1=sm.add_constant(x)
logit_model=sm.Logit(y,x1)
result=logit_model.fit()
print(result.summary())
```

G-statistics = -2*(LL-Null-Log-Likelihood)
=-593.33-(-402.66)
-190.67*(-2)
=381.34
  here number of independent value is 2
here wald statistics is p>|z| that is 0.000<0.05 i.e we reject our null assumption that coefficient of dependent variable is 0. ie our model is significant .
